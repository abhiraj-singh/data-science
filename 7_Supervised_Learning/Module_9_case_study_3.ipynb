### 2. Use LabelEncoder to encode the target variable in to numerical form and split the data such that 20% of the data is set aside for testing.

```python
## apply get dummies function for encoding
df_result = pd.get_dummies(df,drop_first=True)
```

```python
df_result.head(10)
```

```python
df_result.rename(columns={"Private_Yes" : 'Private'},inplace=True)
```

```python
df_result.columns
```

```python
y_col = ["Private"]
x_col = ['Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc', 'F.Undergrad',
       'P.Undergrad', 'Outstate', 'Room.Board', 'Books', 'Personal', 'PhD',
       'Terminal', 'S.F.Ratio', 'perc.alumni', 'Expend', 'Grad.Rate']
```

```python
X = df_result[x_col]
Y = df_result[y_col]
```

```python
Y = np.squeeze(Y)
```

```python
## importing test train split package
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20,random_state=48, stratify = Y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### 3. Fit a linear svm from scikit learn and observe the accuracy.
[Hint: Use Linear SVC]

```python
## Importing Logistic regression Library
from sklearn.svm import LinearSVC
```

```python
lsvc = LinearSVC(random_state = 48, max_iter = 3000, penalty='l2',C=1)
```

```python
lsvc.fit(X_train, y_train)
```

```python
y_pred = lsvc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(f"Accuracy score of Model on Training Data = {round(lsvc.score(X_train, y_train),2)} \n")
```

```python
print(f"Accuracy score of Model on Test Data = {round(accuracy_score(y_test,y_pred),2)} \n\n")
```

```python
print(classification_report(y_test, y_pred))
```

### 4. Preprocess the data using StandardScalar and fit the same model again and observe the change in accuracy.
[Hint: Refer to scikitlearn’s preprocessing methods]

```python
## Apply Standardization 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(X_train)
```

```python
# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#### Fitting the model again

```python
lsvc_2 = LinearSVC(random_state = 48, max_iter = 3000, penalty='l2',C=1)
```

```python
lsvc_2.fit(X_train, y_train)
```

```python
y_pred = lsvc_2.predict(X_test)
```

```python
print(f"Accuracy score of Model on Training Data = {round(lsvc_2.score(X_train, y_train),2)} \n")
```

```python
print(f"Accuracy score of Model on Test Data = {round(accuracy_score(y_test,y_pred),2)} \n\n")
```

```python
print(classification_report(y_test, y_pred))
```

### Fitting the model again after Standardization did increased the accuracy of model.

### 5. Use scikit learn’s gridsearch to select the best hyperparameter for a non-linear SVM, identify the model with best score and its parameters.
[Hint: Refer to model_selection module of Scikit learn]

```python
## importing the non-linear model
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C':[1,1.5,2],
    'kernel':['linear','poly', 'rbf','sigmoid'],
    'gamma':['scale','auto'],
    'max_iter':[100,500,1000,5000],
    'decision_function_shape': ['ovo','ovr']
}
```

```python
svc = SVC(random_state = 48)
```

```python
grid = GridSearchCV(estimator=svc, param_grid=param_grid, cv = 5)
```

```python
grid_results = grid.fit(X_train, y_train)
```

```python
grid_results.cv_results_
```

```python
grid_results.best_params_
```

### Bulilding Model again with best parameters

```python
svc_final = SVC( C= 1.5,
                 decision_function_shape = 'ovo',
                 gamma ='scale',
                 kernel = 'linear',
                 max_iter = 500,
                 random_state = 48)
```

```python
svc_final.fit(X_train, y_train)
```

```python
y_pred = svc_final.predict(X_test)
```

```python
print(f"Accuracy score of Model on Training Data = {round(svc_final.score(X_train, y_train),2)} \n")
```

```python
print(f"Accuracy score of Model on Test Data = {round(accuracy_score(y_test,y_pred),2)} \n\n")
```

```python
print(classification_report(y_test, y_pred))
```
