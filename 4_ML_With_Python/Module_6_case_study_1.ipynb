## 2. Data Manipulation:
### a. Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.
[Hint: Use sum method with appropriate axis]
### b. Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states.

```python
df.head()
```

```python
col_list = ["No. of Inmates benefitted by Elementary Education","No. of Inmates benefitted by Adult Education","No. of Inmates benefitted by Higher Education","No. of Inmates benefitted by Computer Course"]
```

```python
df["total_benefitted"] =  df[col_list].sum(axis=1)
```

```python
df.head()
```

```python
df.loc['totals']= df.sum(numeric_only=True, axis=0)
```

```python
df.tail()
```

```python
df.at['totals',"YEAR"] = np.nan
```

```python
df.tail()
```

## 3. Plotting: 
### a. Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?
[Hint: Use bar method of pyplot]
### b. Make a pie chart that depicts the ratio among different modes of benefits.
[Hint: Use pie method of pyplot]

```python
## Bar Plot
fig = plt.figure(figsize = (35, 10))
plt.bar(df["STATE/UT"][:35],df["total_benefitted"][:35],width = 0.4)
plt.xlabel("STATE/UT")
plt.ylabel("Total Inmates Benefitted")
plt.title("Inmates Benifitted in different courses")
plt.xticks(rotation=90)
plt.show()
```

```python
df_pie = df.loc["totals"][col_list]
```

```python
df_pie
```

```python
# Creating pie plot
fig = plt.figure(figsize =(15, 10))
wp = { 'linewidth' : 1, 'edgecolor' : "black" }
plt.pie(df_pie["totals"], shadow = True, startangle=90,wedgeprops = wp,labels = df_pie["index"],autopct='%1.1f%%')

# show plot
plt.show()
```
