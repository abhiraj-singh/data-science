```python
## Importing the libraries

import pandas as pd
import numpy as np
```

```python
## Loading the CSV files

books = pd.read_csv("./datasets/BX-Books.csv", encoding="latin-1")
users = pd.read_csv("./datasets/BX-Users.csv", encoding="latin-1")
ratings = pd.read_csv("./datasets/BX-Book-Ratings.csv", encoding="latin-1", nrows=10000)
```

```python
## Renaming the colmns of dataframe
books.rename(columns = {'book_title':'title', 'book_author':'author', 'year_of_publication':'year', 'publisher':'publisher'}, inplace=True)
users.rename(columns = {'User-ID':'user_id', 'Location':'location', 'Age':'age'}, inplace=True)
ratings.rename(columns = {'User-ID':'user_id', 'Book-Rating':'rating'}, inplace=True)
```

### Getting the list of user who have done alteast 1 rating on books

```python
x = ratings['user_id'].value_counts() > 0
y = x[x].index
print(y.shape)
ratings = ratings[ratings['user_id'].isin(y)]
```

### Merging the rating with books based on ISBN number into single dataframe

```python
rating_with_books = ratings.merge(books, on='isbn')
rating_with_books.head()
```

### Extracting the list of books which has been rated alteast more than once

```python
number_rating = rating_with_books.groupby('title')['rating'].count().reset_index()
number_rating.rename(columns= {'rating':'number_of_ratings'}, inplace=True)
```

```python
final_rating = rating_with_books.merge(number_rating, on='title')
final_rating.shape
```

```python
final_rating = final_rating[final_rating['number_of_ratings'] >= 2]
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values="rating")
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

### Importing for building recommendation based on NearestNeighbour algorithm

```python
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
book_sparse = csr_matrix(book_pivot)
```

```python
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1))
```

### Best Recommendation for user

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```
