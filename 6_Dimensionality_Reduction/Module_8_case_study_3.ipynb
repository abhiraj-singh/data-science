### apply get dummies function for encoding to convert diagnosis column into an integer column

```python
## apply get dummies function for encoding
df_result = pd.get_dummies(df,drop_first=True)
```

```python
df_result.rename(columns={"diagnosis_M":"diagnosis"},inplace=True)
```

```python
df_result.head()
```

```python
## Generating headmap for determining correlation
plt.figure(figsize = (24, 18))
sns.heatmap(data = df_result.corr(), annot=True, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(20, 220, n=20),
    square=True)
plt.show()
```

#### there are lof of correlated column which can be reduced with dimentionality reduction

```python
x_col = df_result.columns.to_list()
x_col.remove('diagnosis')
```

```python
y_col = ["diagnosis"]
```

```python
X = df_result[x_col]
Y = df_result[y_col]
```

```python
Y = np.squeeze(Y)
```

```python
## importing test train split package
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20,random_state=48, stratify = Y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
## Creating copy of test train split before doing PCA
X_train_orig = X_train
X_test_orig = X_test
y_train_orig = y_train
y_test_orig = y_test
```

## Using PCA decomposition method for dimentionality reduction

```python
## Apply Standardization before applying PCA transformation
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# Fit on training set only.
scaler.fit(X_train)
```

```python
# Apply transform to both the training set and the test set.
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## using PCA method
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 1)
```

```python
## Fitting the tranining data
pca.fit(X_train)
```

```python
## Applying PCA Transformation to both traning and test dataset
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
X_train.shape
```

```python
## Importing Logistic regression Library
from sklearn.linear_model import LogisticRegression
```

```python
## Creating instance of Logistic Regression
lr_pca = LogisticRegression(fit_intercept=True,
                        multi_class='auto',
                        penalty='l2', #ridge regression
                        solver='saga',
                        max_iter=10000,
                        C=50)
```

```python
lr_pca.fit(X_train, y_train)
```

```python
## Training Score
round(lr_pca.score(X_train, y_train),3)
```

```python
## Testing Score
round(lr_pca.score(X_test, y_test),3)
```

```python
from sklearn import metrics
```

```python
y_pred = lr_pca.predict(X_test)
```

```python
metrics.confusion_matrix(y_true=y_test, 
                         y_pred = y_pred, 
                        labels = lr_pca.classes_)
```

```python
print(f"Accuracy score of Model = {round(metrics.accuracy_score(y_test,y_pred),3)}")
```
